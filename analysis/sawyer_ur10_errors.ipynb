```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

DATA_PATH = '../../../data/processed'
```

```python
# Bases processadas por fase, participante e condição
df_eda_p1 = pd.read_csv(f'{DATA_PATH}/eda_normalizado_p1v4.csv')
df_eda_p2 = pd.read_csv(f'{DATA_PATH}/eda_normalizado_p2v4.csv')

# Bases processadas por fase e janelas de 40s
df_pi2905_sawyer = pd.read_excel(f'{DATA_PATH}/PI-2905/40s_overlap50_Sawyer_agg_order.xlsx')
df_pi2905_ur10 = pd.read_excel(f'{DATA_PATH}/PI-2905/40s_overlap50_UR10_agg_order.xlsx')
```

```python
df_eda_p1.shape
```

```python
df_pi2905_sawyer.shape
```

```python
df_eda_p1['Subject'] = df_eda_p1['participant'].apply(lambda p: int(p[-2:]))
df_eda_p1 = df_eda_p1.rename({'condition': 'Condition'}, axis=1)
df_eda_p1[['participant', 'Subject']]
```

```python
df_p1 = pd.merge(left = df_eda_p1[['Subject', 'Condition', 'errors']], right = df_pi2905_sawyer, how="right", on=["Subject", "Condition"])
df_p1['origin'] = 'sawyer'
df_p1
```

```python
df_p1.to_csv('sawyer_errors.csv')
```

```python
df_eda_p2['Subject'] = df_eda_p2['participant'].apply(lambda p: int(p[-3:]))
df_eda_p2 = df_eda_p2.rename({'condition': 'Condition', 'secondary_task': 'Secondary_task'}, axis=1)
df_eda_p2[['participant', 'Subject']]
```

```python
df_p2 = pd.merge(left = df_eda_p2[['Subject', 'Condition', 'Secondary_task', 'errors']], right = df_pi2905_ur10, how="right", on=["Subject", "Condition", "Secondary_task"])
df_p2['origin'] = 'ur10'
df_p2
```

```python
df_p2.to_csv('ur10_errors.csv')
```
